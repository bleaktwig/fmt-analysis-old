```python
def plot(x, y, err, step, xlabel, ylabel):
    fig, ax = plt.subplots()

    # plotting
    plt.scatter(x, y)
    plt.errorbar(x,y,yerr=err)
    
    # additional setup
    fig.set_size_inches(plot_size)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.tight_layout
    
    return
```
